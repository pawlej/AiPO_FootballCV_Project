# AiPO projekt — lokalna inferencja pełnego pipeline'u

Notebook uruchamia kompletny system: detekcja YOLO, ByteTrack, przypisanie drużyn, przypisanie piłki, posiadanie, podania, prędkość/dystans i zapis nagrania wynikowego.


```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT / "src"))
PROJECT_ROOT
```

## Konfiguracja lokalnych ścieżek


```python
VIDEO_PATH = PROJECT_ROOT / "content" / "example_match.mp4"
MODEL_PATH = PROJECT_ROOT / "models" / "best.pt"
OUTPUT_DIR = PROJECT_ROOT / "outputs"

CONFIDENCE = 0.30
MAX_FRAMES = 300  # na start zostaw mały fragment; potem ustaw None
STRIDE = 1
```

```python
assert VIDEO_PATH.exists(), f"Nie znaleziono nagrania: {VIDEO_PATH}"
assert MODEL_PATH.exists(), f"Nie znaleziono modelu: {MODEL_PATH}"
```

## Uruchom pełny pipeline


```python
from aipo_football_cv.pipeline import run_pipeline

result = run_pipeline(
    video_path=VIDEO_PATH,
    model_path=MODEL_PATH,
    output_dir=OUTPUT_DIR,
    confidence=CONFIDENCE,
    max_frames=MAX_FRAMES,
    stride=STRIDE,
    enable_view_transform=False,
)

result
```

## Podejrzyj zapisane statystyki


```python
import pandas as pd

team_stats = pd.read_csv(result.team_stats_path)
player_stats = pd.read_csv(result.player_stats_path)

display(team_stats)
display(player_stats.head(20))
```
